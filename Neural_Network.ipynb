Performing forward and backward pass for a 2 layer neural network by using weights and biases to find output

```python
np.random.seed(0)
input=[[1.9,2.8,5.7],[3.2,2.7,4.1],[5.1,7.6,5.9]]
derival=[[0.5,-0.7],[0.9,0.7],[-0.2,0.5]]
class Layer_Dense:
  def __init__(self,n_inputs,n_neurons):
    self.weights=0.1*np.random.randn(n_inputs,n_neurons)
    self.biases=np.zeros((1,n_neurons))
  def forward(self,inputs):
    self.input=inputs
    self.output=np.dot(inputs,self.weights)+self.biases
    return self.output
  def backward(self,derival,learning_rate=0.1):
    self.derweights=np.dot(np.array(self.input).T,derival)
    self.derbiases=np.sum(derival,axis=0,keepdims=True)
    self.derinputs=np.dot(derival,self.weights.T)
    self.weights-=learning_rate*self.derweights
    self.biases-=learning_rate*self.derbiases
    return self.derinputs
layer1=Layer_Dense(3,5)
layer2=Layer_Dense(5,2)
layer1.forward(input)
layer2.forward(layer1.output)
print(layer1.output)
print(layer2.output)
derinputs_layer2=layer2.backward(derival)
derinputs_layer1=layer1.backward(derinputs_layer2)
print(derinputs_layer1)
```

Performing Rectified Linear Unit Activation to the first layer then performing back propagation

```python
input=[[1.9,2.8,5.7],[3.2,2.7,4.1],[5.1,7.6,5.9]]
output=[]
class Activation_ReLU:
  def forward(self,input):
    self.input=input
    self.output=np.maximum(0,input)
    return self.output
  def backward(self,derival):
    self.derinputs=derival.copy()
    self.derinputs[self.input<=0]=0
    return self.derinputs
layer1=Layer_Dense(3,5)
activation1=Activation_ReLU()
layer1_output=layer1.forward(input)
activation1_output=activation1.forward(layer1_output)
layer2=Layer_Dense(5,2)
layer2_output=layer2.forward(activation1_output)
print(activation1_output)
derival = np.random.randn(*layer2_output.shape)
derinputs_layer2=layer2.backward(derival)
derinputs_activation1=activation1.backward(derinputs_layer2)
derinputs_layer1=layer1.backward(derinputs_activation1)
print(derinputs_layer1)
```

Performing forward pass using ReLU activation function then passing that layer to softmax activation function and then performing backpropagation

```python
input=[[1.9,2.8,5.7],[3.2,2.7,4.1],[5.1,7.6,5.9]]
class Activation_Softmax:
  def forward(self,inputs):
    exp_values=np.exp(inputs-np.max(inputs,axis=1,keepdims=True))
    probabilities=exp_values/np.sum(exp_values,axis=1,keepdims=True)
    self.output=probabilities
    return self.output
  def backward(self,derival,y_true):
    samples=len(derival)
    if len(y_true.shape)==2:
      y_true=np.argmax(y_true,axis=1)
    self.derinputs=derival.copy()
    self.derinputs=self.derinputs/samples
    return self.derinputs
dense1=Layer_Dense(3,4)
activation1=Activation_ReLU()
dense2=Layer_Dense(4,4)
activation2=Activation_Softmax()
dense1.forward(input)
activation1.forward(dense1.output)
dense2.forward(activation1.output)
activation2.forward(dense2.output)
print(activation2.output)
derinputs_layer2=layer2.backward(derival)
derinputs_layer1=layer1.backward(derinputs_layer2)
print(derinputs_layer1)
```

Manual Computation of categorical cross entropy loss for a single prediction by summing the negative log of predicted probabilities weighted by target labels

```python
import math
softmax_output=[0.5,1.6,3.1]
target_output=[1,0,0]
loss=-(math.log(softmax_output[0])*target_output[0]+
       math.log(softmax_output[1])*target_output[1]+
       math.log(softmax_output[2])*target_output[2])
print(loss)
```

Performing forward pass using ReLU for 1 layer and giving that as the input for the other layer that has softmax activation function the caluculating loss using categorical cross entropy and then performing backpropagation in a 2 layer neural network

```python
class Loss:
  def calculate(self,output,y):
    sample_losses=self.forward(output,y)
    data_loss=np.mean(sample_losses)
    return data_loss
class Loss_CategoricalCrossentropy(Loss):
  def forward(self,y_pred,y_true):
    samples=len(y_pred)
    y_pred_clipped=np.clip(y_pred,1e-7,1-1e-7)
    if len(y_true.shape)==1:
      correct_confidences=y_pred_clipped[range(samples),y_true]
    elif len(y_true.shape)==2:
      correct_confidences=np.sum(y_pred_clipped*y_true)
    negative_log_likelihood=-np.log(correct_confidences)
    return negative_log_likelihood
  def backward(self,y_pred,y_true):
    samples=len(y_pred)
    labels=len(y_pred[0])
    if len(y_true.shape)==1:
      y_true=np.eye(labels)[y_true]
    derinputs=(y_pred-y_true)/samples
    return derinputs
layer1=Layer_Dense(3,5)
activation1=Activation_ReLU()
layer2=Layer_Dense(5,2)
activation2=Activation_Softmax()
loss_function=Loss_CategoricalCrossentropy()
layer1_output=layer1.forward(input)
activation1_output=activation1.forward(layer1_output)
layer2_output=layer2.forward(activation1_output)
softmax_output=activation2.forward(layer2_output)
loss=loss_function.calculate(softmax_output,np.array(target_output))
derinputs_layer2=layer2.backward(derival)
derinputs_layer1=layer1.backward(derinputs_layer2)
print(derinputs_layer1)
print("Loss: ",loss)
```

Performing forward pass using sigmoid function in a 2 layer neural network and then performing backpropagation

```python
class Activation_Sigmoid:
  def forward(self,input):
    self.inputs=input
    self.output=1/(1+np.exp(-input))
    return self.output
  def backward(self,derival):
    self.derinputs=derival*(self.output*(1-self.output))
    return self.derinputs
layer1=Layer_Dense(3,2)
activation1=Activation_Sigmoid()
layer1_output=layer1.forward(input)
activation1_output=activation1.forward(layer1_output)
layer2=Layer_Dense(2,2)
layer2_output=layer2.forward(activation1_output)
print(activation1_output)
derival = np.random.randn(*layer2_output.shape)
derinputs_layer2=layer2.backward(derival)
derinputs_activation1=activation1.backward(derinputs_layer2)
derinputs_layer1=layer1.backward(derinputs_activation1)
print(derinputs_layer1)
```
